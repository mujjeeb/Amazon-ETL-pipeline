```python
import pandas as pd
```

```python
df = pd.read_csv('amazon_raw_data.csv')
```

```python
# Strip the product columns of '/n'
df['product_details'] = df['product_details'].astype(str)
df['product_details'] = df['product_details'].str.replace('\n', ' ')
df['product_details']
```

```python
## Cleaning the product details COULMN. Please note the SKU is represented as ASIN by amazon
product_sku = []
rows_to_delete = []
for index, row in df.iterrows():
    
    if 'ASIN' in row['product_details']:
        # Extract ASIN 
        asin = row['product_details'].split('ASIN')[1].split()[1]
        product_sku.append(asin)
    
    else:
        # Mark rows for deletion if 'ASIN' is not found
        rows_to_delete.append(index)
        
        
print(len(product_sku))



# Delete rows marked for deletion
df.drop(rows_to_delete, inplace=True)

# Create a new 'ASIN' column and assign the extracted ASIN values
df['product_sku'] = product_sku

# Reset the index after deleting rows
df.reset_index(drop=True, inplace=True)

df.drop('product_details', axis=1, inplace=True)
```

```python
df
```

```python
 # Checking if there are values that are not ASIN Numbers
    df['product_sku'].unique()
```

```python
unwanted_words = ['Item', 'Customer', 'Batteries',
       'Manufacturer', 'Country', 'Release']
```

```python
# Remove rows that aren't ASIN numbers
condition = ~df['product_sku'].isin(unwanted_words)
filtered_df = df[condition]
```

```python
filtered_df.shape
```

```python
type(filtered_df['price'])
```

```python
# Convert the price column to float
filtered_df['price'] = filtered_df['price'].astype(float)
```

```python
# Extracting the date of the review and country of the review from the column 'date_and_country_of_review'
# Define a regular expression pattern to extract the date and country
pattern = r'Reviewed in (.*?) on (.*?)$'

# Use str.extract() to extract the date and country into separate columns
filtered_df[['country', 'date']] = filtered_df['date_and_country_pf_review'].str.extract(pattern)

filtered_df.drop('date_and_country_pf_review', axis=1, inplace=True)
```

```python
# Remove 'the' from the 'Text' column
filtered_df['country'] = filtered_df['country'].str.replace('the', '', case=False)
```

```python
# Convert date column to datetime format with month day year format

filtered_df['date'] = pd.to_datetime(filtered_df['date'], format='%B %d, %Y')
```

```python
filtered_df
```

```python
# Remove the "out of 5 stars"  strings
filtered_df['rating_elements'] = filtered_df['rating_elements'].str.replace(' out of 5 stars', '')
```

```python
filtered_df['rating_elements'] = filtered_df['rating_elements'].str.replace('<span class="a-icon-alt">', '')
```

```python
filtered_df['rating_elements'] = filtered_df['rating_elements'].str.replace('</span>', '')
```

```python
filtered_df['rating_elements'].unique()
```

```python
# Convert the rating_elements column to float
filtered_df['rating_elements'] = filtered_df['rating_elements'].astype(float)
```

```python
filtered_df['rating_elements']
```

```python
# Rename the 'OldColumnName' to 'NewColumnName'
filtered_df.rename(columns={'rating_elements': 'rating'}, inplace=True)
```
